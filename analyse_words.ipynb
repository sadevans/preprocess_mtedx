```python


# Создание словаря для подсчета уникальных слов для каждого значения ключа "wav"
word_counts = {}
word_pos_counts = {'NOUN': {}, 'VERB': {}, 'ADJECTIVE': {}}

for entry in data:
    wav = entry['wav']
    word = entry['word']
    
    # Подсчет уникальных слов для каждого значения ключа "wav"
    if wav not in word_counts:
        word_counts[wav] = {}
    if word in word_counts[wav]:
        word_counts[wav][word] += 1
    else:
        word_counts[wav][word] = 1
    
    # Разбиение слов на части речи и подсчет
    # tokens = word_tokenize(word, language="russian")
    document = nlp(word)

    for token in document:
    # pos_tags = pos_tag(tokens)
    # for token, pos_tag in pos_tags:
        pos_tag = token.pos_
        lemma = token.lemma_
        if pos_tag.startswith('N'):  # Существительное
            if lemma in word_pos_counts['NOUN']:
                word_pos_counts['NOUN'][lemma] += 1
            else:
                word_pos_counts['NOUN'][lemma] = 1
        elif pos_tag.startswith('V'):  # Глагол
            if lemma in word_pos_counts['VERB']:
                word_pos_counts['VERB'][lemma] += 1
            else:
                word_pos_counts['VERB'][lemma] = 1
        elif pos_tag.startswith('ADJ'):  # Прилагательное
            if lemma in word_pos_counts['ADJECTIVE']:
                word_pos_counts['ADJECTIVE'][lemma] += 1
            else:
                word_pos_counts['ADJECTIVE'][lemma] = 1


```

```python
# Вывод результатов подсчета
print(word_counts)
print(word_pos_counts)
```
